# Customer Personality Data Analysis

# Importing Dataset

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("marketing_campaign.csv", sep="\t")
```

```python
data.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Classification based on Education Status of the clients

```python
l,a = list(set(data["Education"])),[]
for i in range(len(l)):
    a.append(data["Education"].to_list().count(l[i]))
    l[i] = l[i]+" ("+str(a[i]*100/len(data["Education"])) + " %)"
plt.pie(a,labels=l, radius=2)
plt.show()
```

### Classification based on the Marital Status of the clients

```python
l,a = list(set(data["Marital_Status"])),[]
for i in range(len(l)):
    a.append(data["Marital_Status"].to_list().count(l[i]))
plt.bar(x=range(len(l)),height = a)
plt.xticks(labels=l,ticks=range(len(l)), rotation=90)
plt.show()
```

### Classification based on the Age Group of Clients

```python
age=2021-data.Year_Birth.to_numpy()
```

```python
l,a = list(set(age)),[]
for i in range(len(l)):
    a.append(list(age).count(l[i]))
plt.figure(figsize=(12, 4))
plt.bar(x=range(0,len(l)*3,3), height = a)
plt.xticks(labels=l,ticks=range(0,len(l)*3,3),rotation=90)
plt.show()
```

### Correlation between Education Level and Kids

```python
sns.displot(data=data,y="Kidhome",x="Education")
```

### Dependency of Education on Income

```python
sns.displot(data=data, x="Education", y="Income")
```
